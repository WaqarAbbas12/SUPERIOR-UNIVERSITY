# Fill Null Values

```python
def FillNA(df):
    for col in df.columns:
        if df[col].dtype == 'object':  
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].mean())

FillNA(df1)
FillNA(df2)
```

```python
# Check null values again.
print("Df1:", df1.isnull().sum().sum())
print("Df2:", df2.isnull().sum().sum())
```

```python
columns = df1.columns
for col in columns:
    if df1[col].dtype != 'object':
        low, high = df1[col].quantile(0.01), df1[col].quantile(0.99)
        outliers = (df1[col] < low) | (df1[col] > high)

```

```python
df1 = df1[~outliers]
```

```python
columns = df2.columns
for col in columns:
    if df2[col].dtype != 'object':
        low, high = df2[col].quantile(0.01), df2[col].quantile(0.99)
        outliers = (df2[col] < low) | (df2[col] > high)


```

```python

df2 = df2[~outliers]
```

```python
# Encoding of categorical columns
def ordinalencoder(df):

    encoder = OrdinalEncoder()
    categorical_cols = df.select_dtypes(include=['object']).columns
    df[categorical_cols] = encoder.fit_transform(df[categorical_cols])
    
    return df

df1_encoded = ordinalencoder(df1)
df2_encoded = ordinalencoder(df2)
```

```python
def apply_z_score(df):
    return df.apply(lambda col: (col - col.mean()) / col.std() if col.dtype != 'object' else col)

df1_normalized = apply_z_score(df1_encoded)
df2_normalized = apply_z_score(df2_encoded)
```

```python
# Features an Target
x = df1.drop(['MSZoning', 'LotFrontage', 'LotArea', 'Street', 'LotShape', 'LandContour', 'SalePrice'], axis=1)
y = df1['SalePrice']
```

```python
df2_features = df2.drop(['MSZoning', 'LotFrontage', 'LotArea', 'Street', 'LotShape', 'LandContour'], axis=1)
```

```python
model = SVR()
model.fit(x, y)
```

```python
predictions = model.predict(df2_features)
print(predictions)
```

```python
output_df = pd.DataFrame({
    'Id': df2['Id'],  
    'SalePrice': predictions
})

output_df.to_csv('submission.csv', index=False)  
```
