# Task 1: LUHN Algorithm

```python
def luhn_algorithm(number):
    digits = [int(i) for i in number]
    last_digit = digits.pop()
    
    total_sum = 0
    for i in range(len(digits)):
        if (len(digits) - i) % 2 == 0:  # Iterate from reverse side.
            digits[i] *= 2

            if digits[i] > 9:
                digits[i] -= 9
                
        total_sum += digits[i]

    total_sum += last_digit

    if total_sum % 10 == 0:
        print(f"{number} is Valid.")
    else:
        print(f"{number} is Invalid.")

luhn_algorithm("5590490222335810")
```

# Task 2: Remove Punctuations without using remove method

```python
def remove_punctuation(user_input):
    punctuations = '''!()-[]{};:'",<>./?@#$%^&*_~'''
    
    # List comprehension to filter out punctuation
    filtered_chars = [char for char in user_input if char not in punctuations]

    result = ""
    for char in filtered_chars:
        result += char
    print("After remove: ", result)
    
remove_punctuation("Hurrah! We have won the match")
```

# Task 3: Alphabetical Sort without using sort method

```python
# Using Bubble Sort Alogrithm
def sort(string):
    ls = list(string.replace(" ",""))
    
    for i in range(len(ls)-1):
        for j in range(len(ls)-1-i):

            if ls[j] > ls[j+1]:
                ls[j], ls[j+1] = ls[j+1], ls[j]
                
    print("Sorted String: ", "".join(ls))

sort("I am Waqar.")
```
